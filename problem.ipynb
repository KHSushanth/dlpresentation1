```python
import numpy as np

# Input and target
x = np.array([2])
y = np.array([4])

# Initialize weight
w = np.array([1.0])
lr = 0.1

# Training (1 iteration for demo)
# Forward pass
y_hat = w * x

# Loss (MSE)
loss = 0.5 * (y - y_hat) ** 2

# Backward pass (gradient)
grad_w = (y_hat - y) * x

# Update weight
w = w - lr * grad_w

print("Updated weight:", w)
```

```python
import numpy as np

# Input image (2x2)
X = np.array([[1, 2],
              [3, 4]])

# Filter (kernel)
W = np.array([[1.0, 0.0],
              [0.0, 1.0]])

lr = 0.1

# Forward pass (convolution)
Z = np.sum(X * W)

# Assume target output = 3
y = 3

# Loss
loss = 0.5 * (y - Z) ** 2

# Backward pass (gradient of filter)
grad_W = (Z - y) * X

# Update filter
W = W - lr * grad_W

print("Updated filter:\n", W)
```

```python
import numpy as np

# Sequence input [1, 2, 3]
X = [1, 2, 3]
y = 4  # target

# Initialize weights
Wx = 0.5   # input weight
Wh = 0.5   # hidden weight
Wy = 0.5   # output weight

h = 0      # initial hidden state
lr = 0.01

# Forward pass
hs = []
for x in X:
    h = np.tanh(Wx * x + Wh * h)
    hs.append(h)

y_hat = Wy * h

# Loss
loss = 0.5 * (y - y_hat) ** 2

# Backward pass (simplified BPTT)
dWy = (y_hat - y) * h

dh = (y_hat - y) * Wy

# Backprop through time
dWx = 0
dWh = 0

for i in reversed(range(len(X))):
    dtanh = (1 - hs[i] ** 2) * dh
    dWx += dtanh * X[i]
    dWh += dtanh * (hs[i-1] if i > 0 else 0)
    dh = dtanh * Wh

# Update weights
Wx -= lr * dWx
Wh -= lr * dWh
Wy -= lr * dWy

print("Updated Wx:", Wx)
print("Updated Wh:", Wh)
print("Updated Wy:", Wy)
```
